```python
import RPi.GPIO as GPIO
import time

# Define pins
LED2GREEN_pin = 24 # BOARD pin 18, BCM pin 24, ( )

def main():
    # pin setup
    # Board pin-numbering scheme  //  BCM(Boardcom chip-specipic pin number) mode
    GPIO.setwarnings(False)
    GPIO.setmode(GPIO.BCM)  
    # GPIO.setmode(GPIO.BCM)     
    # GPIO.setmode(GPIO.BOARD)     




    #set pin to output mode, and set level is high 
    GPIO.setup(LED2GREEN_pin, GPIO.OUT, initial = GPIO.HIGH)  # DDRn     
     

    print("starting demo now!!")
    curr_velue=GPIO.HIGH  # setup 
    try:
        while True:
            #filp the output every 0.25 sec
            time.sleep(0.25)
            print("outputting {} to pin {}".format(curr_velue, LED2GREEN_pin))  #       
            GPIO.output(LED2GREEN_pin, curr_velue)
            curr_velue ^= GPIO.HIGH  # ( ) 
    finally:
        GPIO.cleanup()
if __name__=='__main__':  #     main    
    main()
```
