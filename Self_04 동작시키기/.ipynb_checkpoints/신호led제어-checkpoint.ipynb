```python
import RPi.GPIO as GPIO
import time

# Define pins
LED2GREEN_pin = 24 # BOARD pin 18, BCM pin 24, (미리 매핑되어있다)

def main():
    # pin setup
    # Board pin-numbering scheme  //  BCM(Boardcom chip-specipic pin number) mode
    GPIO.setwarnings(False)
    GPIO.setmode(GPIO.BCM)  
    # GPIO.setmode(GPIO.BCM)  미리 매핑되어있는 핀번호로 제어
    # GPIO.setmode(GPIO.BOARD) 물리적 보드 번호로 직접 제어




    #set pin to output mode, and set level is high 
    GPIO.setup(LED2GREEN_pin, GPIO.OUT, initial = GPIO.HIGH)  # DDRn과 같이 핀의 입출력 모드를 결정
     

    print("starting demo now!!")
    curr_velue=GPIO.HIGH  # setup의 초깃값
    try:
        while True:
            #filp the output every 0.25 sec
            time.sleep(0.25)
            print("outputting {} to pin {}".format(curr_velue, LED2GREEN_pin))  # 현제 어떤 핀에서 어떤 출력이 나가는지 일려줌
            GPIO.output(LED2GREEN_pin, curr_velue)
            curr_velue ^= GPIO.HIGH  # 출력값(현제 상태) 반전
    finally:
        GPIO.cleanup()
if __name__=='__main__':  # 해당 코드의 엔트리 포인트로 main이 있다면 해당 함수를 실행한다
    main()
```
