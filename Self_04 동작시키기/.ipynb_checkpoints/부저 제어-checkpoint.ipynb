```python
import RPi.GPIO as GPIO
import time
GPIO.setwarnings(False)

# pin definition
BEEP_pin = 6

def main():
    # Pin setup
    # Board pin-numbering scheme
    GPIO.setmode(GPIO.BCM)
    # set pin as an output pin with optional initial state of HIGH
    GPIO.setup(BEEP_pin, GPIO.OUT, initial = GPIO.LOW)
    
    print("Starting demo now!!")
    curr_velue = GPIO.HIGH
    
    try:
        while True:
            time.sleep(1)
            #Toggle the output every second
            GPIO.output(BEEP_pin, curr_velue)
            curr_velue ^= GPIO.HIGH
    finally:
        GPIO.cleanup()
        
if __name__ == '__main__':
    main()
```

```python
# 버튼 누를 때 부저가 5초 동안 울리기

import RPi.GPIO as GPIO
import time
GPIO.setwarnings(False)

# pin definition
BEEP_pin = 6
btn1 = 8
btn2 = 7

def BEEP_call_back1(cnt):
    if cnt != 0:
        cnt = 0
        
    else:
        cnt = 1  
        
        
def BEEP_call_back_timer(time):
    
    if cnt != 0:
        time.sleep(time)
        cnt = 0

        
    else: 
        time.sleep(time)
        cnt = 1

        
def main():
    cnt = 0
    #curr_velue = GPIO.HIGH
    # Pin setup
    # Board pin-numbering scheme
    GPIO.setmode(GPIO.BCM)
    # set pin as an output pin with optional initial state of HIGH
    GPIO.setup(BEEP_pin, GPIO.OUT, initial = GPIO.LOW)
    GPIO.setup(btn1, GPIO.IN)  # default : 1
    GPIO.setup(btn2, GPIO.IN)
    GPIO.add_event_detect(btn1,GPIO.RISING, callback=BEEP_call_back1, bouncetime=10)
#     GPIO.add_event_detect(btn2,GPIO.RISING, callback=BEEP_call_back1, bouncetime=10)
#     GPIO.add_event_detect(btn1,GPIO.RISING, callback=BEEP_call_back1(cnt), bouncetime=10)
#     print(check2)
    print(cnt)
#     GPIO.add_event_detect(btn2,GPIO.RISING, callback=BEEP_call_back2, bouncetime=10)
    print("Starting demo now!!")
   
    try:
        while True:
            
          
            if cnt != 0:
                GPIO.output(BEEP_pin, 1)
                BEEP_call_back_timer(1)
                print(cnt)
                
            else:
                GPIO.output(BEEP_pin, 0)
                BEEP_call_back_timer(1)
                print(cnt)
                
            pass
    finally:
        GPIO.cleanup()
        
if __name__ == '__main__':

    main()
```
